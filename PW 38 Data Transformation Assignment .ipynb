### Q1. Handle Missing Values (Basic Cleaning)
#### Fill missing values using the following:
- Age → Median
- Salary → Mean
- Hours_Worked_Per_Week → Median
- Performance_Score → Mean
- Display the dataset after handling missing values. 


#### Answer:

```python
import pandas as pd
```

```python
# 1. Loading dataset
df = pd.read_csv('employee_productivity_dataset.csv')
```

```python
# 2. Listing missing values in dataset
missing_values = df.isnull().sum()
print('Missing Values :', missing_values)
```

```python
# 3. Filling respective missing values as instructed
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].mean())
```

```python
# 4. Conferming filled missing values
check_missing_values = df.isnull().sum()
print('Missing Values after cleaning :', check_missing_values)
```

```python
# Displaying cleaned dataset
df
```

### Q7. Build Preprocessing Pipeline
### Create a pipeline that:
- Applies encoding to categorical columns
- Applies scaling to numerical columns
### Use:
- ColumnTransformer
- Pipeline


#### Answer:

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# 1. Defineing column groups
categorical_cols = ['Gender', 'Department', 'Work_Mode', 'Location']
numerical_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Performance_Score', 'Projects_Completed']
```

```python
# 2. Createing transformers
categorical_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')
numerical_transformer = StandardScaler()
```

```python
# 3. Combineing using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)])
```

```python
# 4. Createing pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor)])
```

### Q8. Apply Pipeline
#### Apply the pipeline on the dataset.
### Display:
- Transformed dataset
- Shape of final dataset

#### Answer:

```python
# 1. Applying pipeline
processed_data = pipeline.fit_transform(df)
```

```python
# 2. Converting to DataFrame for better display
processed_df = pd.DataFrame(processed_data)
```

```python
# 3. Display transformed dataset
processed_df.head()
```

```python
# Displaying shape of final dataset
print("Final dataset shape:", processed_df.shape)
```

### Q9. Conceptual Question
### Why is scaling important in Python?

#### Answer :
Scaling is important because it brings all features to a similar range.
- It prevents large values (like Salary) from dominating smaller ones (like Age)
- It helps machine learning models work more accurately
- It improves speed and performance of algorithms

In simple terms, scaling makes the data fair for all features.

### Q10. Conceptual Question
### Why do we convert categorical data into numerical form?

#### Answer :
We convert categorical data into numerical form because machine learning models cannot understand text data directly.
- Models work with numbers, not words (e.g., "Male", "HR")
- Converting categories into numbers makes the data usable
- It helps models process and learn patterns from the data

In simple terms, encoding translates text into numbers so machines can understand it.
