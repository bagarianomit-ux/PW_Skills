### Q2. Explain the difference between List and Tuple in Python.

#### Answer:

In Python, both lists and tuples are used to store collections of elements, but they differ mainly in terms of mutability and usage. A list is mutable, which means its elements can be modified, added, or removed after creation, making it suitable for dynamic data. In contrast, a tuple is immutable, meaning once it is created, its elements cannot be changed, which makes it more secure and faster in execution. Lists are defined using square brackets [], while tuples use parentheses (). Due to their flexibility, lists have more built-in methods, whereas tuples have limited functionality. 

#### Conclusion:
Lists are flexible and useful for modifying data, while tuples are more efficient and secure for storing constant data.

### Q5. Write a Python program to check whether a number is even or odd using conditional statements.

#### Answer:

In this program, we take a number as input and use a conditional statement to check whether it is even or odd. A number is even if it is divisible by 2, otherwise it is odd.

````python
# Using a fixed value to avoid input issues
num = 10
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
````

### Q6. Write a program to print numbers from 1 to 10 using a loop.

#### Answer:

In this program, a loop is used to print numbers from 1 to 10. The ```for``` loop in Python is commonly used when the sequence is known in advance. Here, the ```range()``` function helps generate numbers in the desired sequence.

````python
# Printing numbers from 1 to 10 using a loop

for Numbers in range(1, 11):
    print(Numbers)
````

### Q7. Create a list of five numbers and print the maximum number from the list.

#### Answer:

In this program, a list of five numbers is created (55,81,94,64,73) and the maximum value is found using the built-in ```max()``` function. This function helps to quickly identify the largest element in a list.

### Q8. Write a Python program to remove duplicate values from a list using a set.

#### Answer:

In this program, a list that contain duplicate values is converted into a set. Since sets do not allow duplicate elements, this automatically removes duplicates.

````python
# Firstly we had original list with duplicate values i.e numbers,
# Then it is converted into set i.e unique_numbers which is then printed to give list without duplicate elements. 

numbers = [1, 2, 3, 2, 4, 7, 5, 6, 1, 3, 4, 8]
unique_numbers = list(set(numbers))
print("List after removing duplicates:", unique_numbers)
````

### Q10. Write a Python program to count how many times a number appears in a list.Example List [2,3,4,2,5,2]

#### Answer:

In this program, a specific number is searched within the given example list. The ```count()``` method is used to find how many times that number appears in the list.

````python
# Given list
numbers = [2, 3, 4, 2, 5, 2]

# Let the number to be counted = 2 so,

num = 2
count = numbers.count(num)
print("Number", num, "appears", count, "times in the list.")
````
