### Q1. Understanding Central Tendency (Easy) A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12]. What is the most representative value of their weekly sales, and why?

#### Answer :

The most representative value can be found using the mean because the sales data does not contain any extreme values.

Mean = (10 + 12 + 11 + 15 + 14 + 13 + 12) / 7 = 87 / 7 = 12.43

Result: Mean sales = 12.43 dozen muffins.

Interpretation: On average, the bakery sells about 12.43 dozen muffins per day.

Conclusion: The mean is the most representative value because it summarizes the overall weekly sales level.

### Q2. Mean in Real Life (Easy) A teacher records the marks of her students in a short quiz: [12, 15, 14, 16, 18, 20, 19]. What is the mean score, and what does it tell us about the class’s performance? 


#### Answer :

The mean score is calculated as the sum of all marks divided by the number of students.

Mean = (12 + 15 + 14 + 16 + 18 + 20 + 19) / 7 = 114 / 7 = 16.29

Result: Mean score = 16.29.

Interpretation: The average student scored approximately 16.29 marks.

Conclusion: The class performed reasonably well, with scores centered around 16 marks.

### Q3. Mode in Real Life (Easy) A store records the shoe sizes sold in one day: [7, 8, 9, 8, 8, 10, 7, 9]. What is the mode, and why is this information useful for the store manager?


#### Answer :

The mode is the value that occurs most frequently.

Shoe sizes: [7, 8, 9, 8, 8, 10, 7, 9]

Mode = 8

Result: Mode = 8.

Interpretation: Size 8 was sold more often than any other size.

Conclusion: The store manager can use this information to maintain sufficient stock of size 8 shoes.

### Q4. Median in Real Life (Medium) A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000]. Why is the median a better measure than the mean in this case? Calculate the median. 


#### Answer :

The median is the middle value when data is arranged in order.

Prices: [`$8,000`, `$9,500`, `$10,200`, `$11,000`, `$50,000`]

Median = `$10,200`

The mean would be heavily influenced by the unusually high value of `$50,000`.

Result: Median = `$10,200`.

Interpretation: Half of the cars cost less than `$10,200` and half cost more.

Conclusion: The median is a better measure because it is not affected by the extreme price of `$50,000`.


### Q5. Dispersion Introduction (Medium) A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40]. What does the range tell us about the variation in the student’s puzzle-solving time? 


#### Answer :

The range measures the spread of the data.

Range = Maximum − Minimum
= 40 − 25
= 15 minutes

Result: Range = 15 minutes.

Interpretation: The student's puzzle-solving time varies by up to 15 minutes between the fastest and slowest days.

Conclusion: The range shows the amount of variation in the student's performance.

### Q6. Range in Action (Medium) A farmer records the weekly weight of harvested apples (kg): [100, 105, 98, 110, 120]. Find the range. How can this help the farmer in planning his packaging?

#### Answer :

The range is calculated as the difference between the highest and lowest weights.

Range = 120 − 98
= 22 kg

Result: Range = 22 kg.

Interpretation: Weekly harvest weights vary by 22 kg.

Conclusion: Knowing the range helps the farmer plan packaging capacity to accommodate fluctuations in harvest volume.

### Q7. Variance for Decision-Making (Medium) Two delivery companies track delivery delays (in minutes). Company A: variance = 6 Company B: variance = 15 Which company is more consistent, and why? 

#### Answer :

Variance measures how spread out data values are from the mean.

`Company A Variance = 6`
`Company B Variance = 15`

Result: Company A has the lower variance.

Interpretation: Delivery delays for Company A vary less from day to day.

Conclusion: Company A is more consistent because lower variance indicates more stable performance.

### Q8. Standard Deviation in Context (Hard) A finance student compares the daily price fluctuations of two cryptocurrencies. Coin A: standard deviation = `$30` Coin B: standard deviation = `$120` Which coin is riskier to invest in, and why? 

#### Answer :

Standard deviation measures the typical distance of values from the mean.

`Coin A Standard Deviation = $30`
`Coin B Standard Deviation = $120`

Result: Coin B has the larger standard deviation.

Interpretation: Coin B experiences much larger price fluctuations.

Conclusion: Coin B is riskier because its prices are more volatile than Coin A's.

### Q9. Combining Measures (Hard) A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410] Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?

#### Answer :

The mean measures average electricity usage, while the standard deviation measures variability.

Data: [400, 420, 390, 450, 410]

Mean = (400 + 420 + 390 + 450 + 410) / 5
= 414 kWh

Sample Standard Deviation ≈ 22.80 kWh

Result:
- Mean = 414 kWh
- Standard Deviation ≈ 22.80 kWh

Interpretation: The family uses about 414 kWh of electricity per month, with moderate variation from month to month.

Conclusion: Electricity consumption is relatively stable around the average value.

### Q10. Practical Application (Hard) A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21]. Find the mean, median, mode, range, and standard deviation. What insights can these measures provide about the player’s scoring performance?


#### Answer :

Data: [15, 18, 20, 22, 25, 17, 19, 21]

Mean = 19.63

Sorted Data: [15, 17, 18, 19, 20, 21, 22, 25]

Median = (19 + 20) / 2 = 19.5

Mode = No mode (all values occur once)

Range = 25 − 15 = 10

Sample Standard Deviation ≈ 3.34

Results:
- Mean = 19.63
- Median = 19.5
- Mode = No mode
- Range = 10
- Standard Deviation ≈ 3.34

Interpretation:
- The player scores about 20 points per game on average.
- The mean and median are very close, indicating a balanced distribution of scores.
- The absence of a mode shows no score occurred more frequently than others.
- The range of 10 points indicates moderate variation.
- The standard deviation of 3.34 suggests fairly consistent scoring performance.

Conclusion:
The player is a reliable scorer whose performance remains relatively consistent across games.
