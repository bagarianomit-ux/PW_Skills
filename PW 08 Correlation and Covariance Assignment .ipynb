### Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

#### Answer :

Covariance is a statistical measure that shows the direction of the relationship between two variables. It indicates whether two variables increase together or move in opposite directions.

The covariance between two variables X and Y is given by:

Cov(X,Y) = Σ[(Xi − X̄)(Yi − Ȳ)] / (n − 1)

Difference between Covariance and Correlation:

| Covariance | Correlation |
|------------|-------------|
| Measures the direction of relationship between two variables. | Measures both the direction and strength of the relationship. |
| Depends on the units of measurement. | Unit-free measure. |
| Values are not bounded. | Values always lie between -1 and +1. |
| Difficult to interpret because its magnitude depends on scale. | Easy to interpret because of its fixed range. |

Therefore, correlation is generally preferred because it provides a standardized measure of association.

### Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

#### Answer :

Covariance indicates how two variables move together.

- **Positive covariance:** Both variables tend to increase or decrease together.

- **Negative covariance:** One variable tends to increase while the other decreases.

- **Zero covariance:** There is no linear relationship between the two variables.

Thus, the sign of covariance tells the direction of the relationship, while its magnitude alone is difficult to interpret.

### Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

#### Answer :

Covariance has the following limitations:

- Its value depends on the units of measurement.
- There is no fixed range, making interpretation difficult.
- It only indicates the direction of the relationship.
- The magnitude cannot be compared across different datasets.

Correlation is preferred because:

- It is independent of the measurement units.
- Its values always lie between -1 and +1.
- It indicates both the strength and direction of the relationship.
- It allows easy comparison between different datasets.

### Q4. Explain the difference between Pearson's correlation coefficient and Spearman's rank correlation coefficient. When would you prefer to use Spearman's correlation?

#### Answer :

**Pearson's Correlation**

- Measures the linear relationship between two variables.
- Uses the actual numerical values.
- Assumes the data is normally distributed.
- Suitable for continuous numerical data.

**Spearman's Rank Correlation**

- Measures the relationship using the ranks of the observations.
- Does not require normally distributed data.
- Suitable for ordinal data or non-linear monotonic relationships.
- Less affected by outliers.

Spearman's correlation is preferred when the data is ordinal, contains outliers, or does not satisfy the assumptions required for Pearson's correlation.

### Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?

#### Answer :

A correlation coefficient of **0.85** indicates a strong positive linear relationship between X and Y.

This means that as X increases, Y also tends to increase.

However, correlation does not imply causation.

A high correlation only shows that the variables are associated. It does not prove that changes in one variable cause changes in the other. The relationship may be influenced by other variables or may simply be coincidental.

### Q6. Using the dataset below, calculate the covariance between X and Y.

X = [2, 4, 6, 8]

Y = [3, 7, 5, 10]

#### Answer :

Formula:

Cov(X,Y) = Σ[(Xi − X̄)(Yi − Ȳ)] / (n − 1)

Mean of X:

X̄ = (2 + 4 + 6 + 8) / 4 = 5

Mean of Y:

Ȳ = (3 + 7 + 5 + 10) / 4 = 6.25

| X | Y | X−X̄ | Y−Ȳ | (X−X̄)(Y−Ȳ) |
|---|---|------|------|-------------|
|2|3|-3|-3.25|9.75|
|4|7|-1|0.75|-0.75|
|6|5|1|-1.25|-1.25|
|8|10|3|3.75|11.25|

Σ[(X−X̄)(Y−Ȳ)] = 19

Covariance = 19 / (4 − 1)

= 19 / 3

= **6.33**

Therefore, the covariance between X and Y is **6.33**.

```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([3, 7, 5, 10])

covariance = np.cov(x, y)[0, 1]
print("Covariance =", covariance)
```

### Q7. Compute the Pearson correlation coefficient between variables A and B.

A = [10, 20, 30, 40, 50]

B = [8, 14, 18, 24, 28]

#### Answer :

The Pearson correlation coefficient measures the strength and direction of the linear relationship between two variables.

In Python, it can be calculated using the `corrcoef()` function.

The Pearson correlation coefficient is:

r ≈ **0.998**

Since the value is very close to +1, there is a very strong positive linear relationship between A and B.

```python
import numpy as np

A = np.array([10, 20, 30, 40, 50])
B = np.array([8, 14, 18, 24, 28])

correlation = np.corrcoef(A, B)[0, 1]
print("Pearson Correlation =", correlation)
```

### Q8. The following table shows heights (in cm) and weights (in kg) of 5 students.

Height = [150, 160, 165, 170, 180]

Weight = [50, 55, 58, 62, 70]

Find the correlation coefficient between Height and Weight.

#### Answer :

The Pearson correlation coefficient is used to measure the relationship between height and weight.

The correlation coefficient is:

r ≈ **0.992**

Since the value is very close to +1, height and weight have a very strong positive correlation.

This means that as height increases, weight also tends to increase.

```python
import numpy as np

height = np.array([150, 160, 165, 170, 180])
weight = np.array([50, 55, 58, 62, 70])

correlation = np.corrcoef(height, weight)[0, 1]
print("Correlation =", correlation)
```

### Q10. Two investment portfolios have the following returns (%) over 5 years. Compute the covariance and correlation coefficient, and interpret whether the portfolios move together.

Portfolio A = [8, 10, 12, 9, 11]

Portfolio B = [6, 9, 11, 8, 10]

#### Answer :

Covariance measures how two variables move together, while correlation measures both the strength and direction of their relationship.

Using the data:

Covariance = **3.00**

Correlation coefficient (r) ≈ **0.986**

Since the correlation coefficient is very close to +1, the two portfolios have a very strong positive relationship.

This indicates that the returns of both portfolios generally move together.

```python
import numpy as np

portfolio_A = np.array([8, 10, 12, 9, 11])
portfolio_B = np.array([6, 9, 11, 8, 10])

covariance = np.cov(portfolio_A, portfolio_B)[0, 1]
correlation = np.corrcoef(portfolio_A, portfolio_B)[0, 1]

print("Covariance =", covariance)
print("Correlation =", correlation)
```
