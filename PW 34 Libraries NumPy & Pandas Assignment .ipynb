### Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.

#### Answer: 
Steps:
1. Check for missing values in each column using `.isnull()`.
2. Count the number of missing values column-wise.
3. Calculate the total missing values in the entire dataset.


````python
# Checking missing values in each column and adding them to find the total number of missing values per column.
missing_per_column = df.isnull().sum()
print(missing_per_column)
````

### Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.

#### Answer:

Steps:

1. Calculate the mean of Quantity and Price columns.
2. Replace missing values `NaN` with their respective mean using `.fillna().`
3. To verify change check missing values in Quantity and Price columns.


````python
# Calculating the mean of Quantity and Price columns and replace missing values with their respective mean using.
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
````

````python
#  verifying changes.
print("Missing values after replacement:")
print(df[['Quantity', 'Price']].isnull().sum())
````

### Q4. Remove all rows where the Product Category or Region column contains missing values.

### Q5. Create a new column called Revenue using NumPy, where: Revenue = Quantity × Price ; Display the updated dataset with the new column.

#### Answer :

steps:

1. Use NumPy to create a new column called **Revenue**.
2. Apply the formula: Revenue = Quantity × Price
3. Add this new column to the dataset.


````python
import numpy as np
````

````python
df = df.dropna(subset=['Product Category', 'Region'])
````

````python
# Createing Revenue column.
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])
````

````python
# Displaying updated dataset
print("Dataset with Revenue column:")
(df.head())
````

### Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

#### Answer:

````python
# Calculateing total revenue using NumPy
total_revenue = np.sum(df['Revenue'])
print("Total Revenue:")
print(total_revenue)
````

### Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

#### Answers:

steps:

1. Group the dataset by Product Category.
2. Round off SUM value up to two decimal places for cleaner value and then print the values for categorycal revanue sum.


````python
# Grouping by Product Category and calculateing total revenue.

category_revenue = df.groupby('Product Category')['Revenue'].sum()

# rounding off up to two decimal points for cleaner value and then printing the values for categorycal revanue sum.

category_revenue = category_revenue.round(2)
print(category_revenue)
````

### Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.

#### Answer:

steps:

1. Short table in Decending order of category revanue.
2. To get top 3 categories with highest revenue print top 3 rows of the shorted table.
3. To get Bottom 3 categories with lowest revenue print last 3 rows of the shorted table.

````python
# Short table in Decending order of category revanue.
sorted_revenue = category_revenue.sort_values(ascending=False)
````

````python
# To get top 3 categories with highest revenue print top 3 rows of the shorted table.
top_3 = sorted_revenue.head(3)
print("Top 3 Categories by Revenue:")
print(top_3)
````

````python
# To get Bottom 3 categories with lowest revenue print last 3 rows of the shorted table.
bottom_3 = sorted_revenue.tail(3)
print("\nBottom 3 Categories by Revenue:")
print(bottom_3)
````

### Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.

#### Answer:

steps:

1. Group the dataset by Region.
2. Calculate the total revenue for each region.
3. Short table in Decending order of region vise revanue.
4. Identify the region with the highest revenue (First row).
5. Identify the region with the lowest revenue (Last row).



````python
# step-1:
region_revenue = df.groupby('Region')['Revenue'].sum().round(2)
print("Revenue by Region:")
print(region_revenue)
````

````python
# step-2:
highest_region = region_revenue.idxmax()
print("Region with Highest Revenue:", highest_region)
````

````python
# step-3:
lowest_region = region_revenue.idxmin()
print("Region with Lowest Revenue:", lowest_region)
````

### Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.

#### Answer:

steps:

1. Converting Date column correctly (day first format) to Convert Date column to datetime formate.
2. Extract the month from the Date column.
3. Group the dataset by month and calculate total monthly revenue.
4. Print month vise total revenue.
5. Calculate the mean, median, and standard deviation of the Revenue column using NumPy.

````python
# step-1:
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
````

````python
# step-2:
df['Month'] = df['Date'].dt.month
````

````python
# step-3:
monthly_revenue = df.groupby('Month')['Revenue'].sum()
````

````python
# step-4:
print("Monthly Revenue:")
print(monthly_revenue)
````

````python
# step-5:
mean_revenue = np.mean(df['Revenue'])
median_revenue = np.median(df['Revenue'])
std_revenue = np.std(df['Revenue'])


print("Mean:", (mean_revenue))
print("Median:", (median_revenue))
print("Standard Deviation:",(std_revenue))
````
