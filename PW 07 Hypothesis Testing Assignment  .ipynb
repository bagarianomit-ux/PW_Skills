### Q1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?

#### Answer :

The null hypothesis (H₀) is a statement that assumes there is no significant difference, effect, or relationship between variables. It is considered true unless there is sufficient evidence to reject it.

The null hypothesis serves as the starting point of hypothesis testing. Statistical tests are used to determine whether the sample data provides enough evidence to reject H₀ in favor of the alternative hypothesis (H₁).

### Q2. What does the significance level (α) represent in hypothesis testing?

#### Answer :

The significance level (α) is the maximum probability of rejecting the null hypothesis when it is actually true. This is known as a Type I error.

The most commonly used significance level is 0.05.

For example, if α = 0.05, there is a 5% risk of rejecting the null hypothesis even though it is true.

### Q3. Differentiate between Type I and Type II errors.

#### Answer :

| Type I Error | Type II Error |
|--------------|---------------|
| Rejecting a true null hypothesis. | Failing to reject a false null hypothesis. |
| Also called a False Positive. | Also called a False Negative. |
| Probability = α (Significance Level). | Probability = β. |
| Example: Concluding a medicine works when it actually does not. | Example: Concluding a medicine does not work when it actually does. |

### Q4. Explain the difference between a one-tailed and two-tailed test. Give an example of each.

#### Answer :

A one-tailed test is used when the alternative hypothesis specifies the direction of the effect (greater than or less than).

Example:

A company claims that a new machine increases production.

H₀: μ = 100

H₁: μ > 100

A two-tailed test is used when the alternative hypothesis only states that there is a difference, without specifying the direction.

Example:

A company claims that the average production is different from 100 units.

H₀: μ = 100

H₁: μ ≠ 100

A one-tailed test checks for an effect in one direction, while a two-tailed test checks for an effect in both directions.

### Q5. A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

#### Answer :

Given:

- Population mean (μ) = 10 minutes
- Sample mean (x̄) = 12 minutes
- Sample standard deviation (s) = 3 minutes
- Sample size (n) = 9
- Significance level (α) = 0.05

Since the population standard deviation is unknown and the sample size is less than 30, a one-sample t-test is used.

Hypotheses:

H₀: μ = 10

H₁: μ ≠ 10

Test Statistic:

t = (x̄ − μ) / (s / √n)

= (12 − 10) / (3 / √9)

= 2 / 1

= 2.00

Degrees of freedom (df) = 9 − 1 = 8

Critical value at α = 0.05 (two-tailed):

t₀.₀₂₅,₈ = ±2.306

Since |2.00| < 2.306, we fail to reject the null hypothesis.

Therefore, there is not enough evidence at the 5% significance level to conclude that the average complaint resolution time is different from 10 minutes.

### Q6. When should you use a Z-test instead of a t-test?

#### Answer :

A Z-test is used when:

- The population standard deviation (σ) is known.
- The sample size is large (generally n ≥ 30).
- The data follows a normal distribution or the sample size is sufficiently large.

A t-test is used when:

- The population standard deviation is unknown.
- The sample size is small (generally n < 30).
- The sample standard deviation is used to estimate the population standard deviation.

### Q7. The productivity of 6 employees was measured before and after a training program. At α = 0.05, test if the training improved productivity.

#### Answer :

Since the productivity of the same employees is measured before and after training, a paired t-test is used.

Before: 50, 60, 58, 55, 62, 56

After: 55, 65, 59, 58, 63, 59

Difference (After − Before):

5, 5, 1, 3, 1, 3

Mean difference (d̄):

= (5 + 5 + 1 + 3 + 1 + 3) / 6

= 3

Sample standard deviation of differences:

s ≈ 1.79

Number of observations:

n = 6

Test statistic:

t = d̄ / (s / √n)

= 3 / (1.79 / √6)

≈ 4.10

Degrees of freedom:

df = 6 − 1 = 5

Critical value for a one-tailed test at α = 0.05:

t₀.₀₅,₅ = 2.015

Since 4.10 > 2.015, we reject the null hypothesis.

Therefore, there is sufficient evidence at the 5% significance level to conclude that the training program improved employee productivity.
