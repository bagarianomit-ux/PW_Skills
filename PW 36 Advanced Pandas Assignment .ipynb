### Q2 Sales Filtering Analysis
### Filter the dataset to display:
1. Orders where Sales > 3500
2. Orders where Region = "East" and Sales > 2500
### Explain what insights you observe.

#### Answer:

````python
# 1. Orders where Sales > 3500

high_sales = df[df['Sales'] > 3500]
high_sales
````

````python
# 2. Orders where Region = "East" and Sales > 2500

east_high_sales = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]
east_high_sales
````

````python
# For batter insights

print("Most repeated high value category overall:")
print(high_sales['Category'].value_counts())

print("Top high value  Category overall:")
print(high_sales['Category'].value_counts().idxmax())

print("Most repeated high value category in east region:")
print(east_high_sales['Category'].value_counts())

print("Top high value Category in east region:")
print(east_high_sales['Category'].value_counts().idxmax())
````

#### Insights:
- The most repeated high-value category overall is Electronics (56 orders), followed by Accessories and Office products.
  
- In the East region, Electronics again dominates with 34 high-value orders, significantly higher than other categories. This shows that the East region has a particularly strong market for Electronics compared to Accessories and Office supplies.

- Electronics is the top-performing category both overall and in the East region

### Q3 Customer Purchase Analysis
### Identify the top 5 customers based on total sales.
(Hint: Use GroupBy with aggregation.)

#### Answer:
##### Steps:
1. Group the dataset by Customer Name
2. Calculate total sales using sum()
3. Sort the values in descending order
4. Select the top 5 customers

````python
# Grouping Customers and calculateing total sales
customer_sales = df.groupby('Customer')['Sales'].sum()
````

````python
# Sorting in descending order and printing top 5 customers.
top_5_customers = customer_sales.sort_values(ascending=False).head(5)
print(top_5_customers)
````

### Q4 Regional Sales Performance
### Calculate the total sales generated by each region.
### Then identify:
- Which region generated the highest revenue?

#### Answer:
##### Steps:
1. Group the dataset by 'Region'
2. Calculate total sales using sum()
3. Compare values to identify the highest-performing region



````python
# Calculateing total sales by region.
region_sales = df.groupby('Region')['Sales'].sum()
print("Total Sales by Region:")
print(region_sales)
````

````python
# Region with highest revenue.
top_region = region_sales.idxmax()
print("Region with Highest Revenue:")
print(top_region)
````

### Q5 Product Performance Analysis

### Q6 Multi-Level Grouping
- Group the dataset by: Region and Category
- Calculate the total sales for each combination.
- What insights can you derive from this analysis?

#### Answers:
##### Steps:
1. Group the dataset by both 'Region' and 'Category' and Calculate total sales using sum().
2. Sum all categories within each region.
3. Insights

````python
# Step 1 :
region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum()
print(region_category_sales)
````

````python
# Step 2 :
region_total_sales = region_category_sales.groupby('Region').sum()
print("Total Sales for Each Region:")
print(region_total_sales)
````

#### Insights:
- The East region generates the highest total sales (232,675), making it the top-performing region overall.
- Across all regions, Electronics is the highest revenue-generating category.
- Office category contributes the least in every region, indicating lower sales volume or demand.


#### Conclusion:
  - Focus on Electronics for maximum revenue.
  - Strengthen strategies in the North region.
  - Maintain strong performance in the East region, especially in Electronics.

### Q7 Pivot Table Creation
### Create a pivot table showing:
1. Rows → Region
2. Columns → Product
3. Values → Total Sales
- Which region sells the most laptops?



#### Answer:

````python
# Createing pivot table.

pivot_table = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)
print(pivot_table)
````

````python
# Find which region sells the most laptops
top_laptop_region = pivot_table['Laptop'].idxmax()
print("Region with Highest Laptop Sales:")
print(top_laptop_region)
````

### Q8 Profitability Analysis
### Calculate the average profit for each product category. And Which category is most profitable?

#### Answers:
##### Steps-
1. Group the dataset by 'Category' and Calculate the average profit using ```.mean()```
2. Identify the most profitable category using ```.idxmax()```

### Q9 Discount Impact Study
### Analyze the relationship between discount and profit.
#### Tasks:
1. Calculate the average profit for each discount level.
2. Identify whether higher discounts lead to lower profit.

#### Answer:

````python
# Group the dataset by 'Discount' and Calculate the average profit.
discount_profit = df.groupby('Discount')['Profit'].mean()
print("Average Profit for Each Discount Level:")
print(discount_profit)
````

````python
# Sorting data for better understanding
print("Sorted (by Discount Level):")
print(discount_profit.sort_index())
````

#### Insights:
- There is a **clear negative relationship between discount and profit**.
- As discount increases, average profit consistently decreases:
- This shows that higher discounts reduce profit margins.

### Q10 End-to-End Data Wrangling Task
### Perform the following workflow:
1. Filter the dataset where Sales > 3000.
2. Group data by Region and Product.
3. Calculate total sales.
4. Reshape the result into a pivot table.
- Explain the business insights obtained from this analysis.

#### Answer:
##### Steps-
1. Filter data where Sales > 3000
2. Group by Region and Product and calculate total sales
4. Reshape the data into a pivot table

````python
# Step 1:
high_value_sales = df[df['Sales'] > 3000]
````

````python
# Step 2:
grouped_data = high_value_sales.groupby(['Region', 'Product'])['Sales'].sum()
````

````python
# Step 3:
pivot_table = grouped_data.unstack()
print(pivot_table)
````

#### Insights:
- Laptops are highly concentrated in the East region, making it the premium market
- Accessories (Mouse, Keyboard, Headphones) perform well across all regions
- Product preference varies by region, indicating need for region-specific strategy

#### Business Value:
- Focus **premium products (Laptops, Phones)** in East
- Promote **balanced product mix** in South
- Target **accessories market** in West
- Improve **high-value product penetration** in North
