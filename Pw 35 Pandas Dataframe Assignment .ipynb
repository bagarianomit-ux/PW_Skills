### Q7. Filtering Data
### Filter the dataset to display:
1. Rows where Region = "East".

#### Answer:

### Q9. Descriptive Statistics
### Generate summary statistics using: describe()
#### From the result identify:
1. Mean sales value
2. Maximum quantity sold
3. Minimum sales value

#### Answer:

```python
# Generateing summary statistics
summary = df.describe()
print(summary)
```

```python
# Mean sales value
mean_sales = summary.loc['mean', 'Sales']
print("Mean Sales Value:", mean_sales)
```

```python
# Maximum quantity sold
max_quantity = summary.loc['max', 'Quantity']
print("Maximum Quantity Sold:", max_quantity)
```

```python
# Minimum sales value
min_sales = summary.loc['min', 'Sales']
print("Minimum Sales Value:", min_sales)
```

### Q10. Aggregation Analysis
### Using the Sales column, calculate:
1. Total sales
2. Average sales
3. Maximum sales
4. Minimum sales

#### Answer: 

```python
# Total sales
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
# Average sales
average_sales = df['Sales'].mean()
print("Average Sales:", average_sales)
```

```python
# Maximum sales
max_sales = df['Sales'].max()
print("Maximum Sales:", max_sales)
```

```python
# Minimum sales
min_sales = df['Sales'].min()
print("Minimum Sales:", min_sales)
```
