### Q1. A die is rolled. What is the probability of getting:
(a) An even number
(b) A number greater than 4

#### Answer :

Sample space = {1, 2, 3, 4, 5, 6}

(a) Even numbers = {2, 4, 6}

Probability of getting an even number

P(Even) = 3/6 = 1/2 = 0.5

(b) Numbers greater than 4 = {5, 6}

Probability of getting a number greater than 4

P(Number > 4) = 2/6 = 1/3 ≈ 0.333

Result:
- P(Even) = 1/2
- P(Number > 4) = 1/3

Conclusion: An even number has a 50% chance of occurring, while a number greater than 4 has a probability of about 33.3%.

### Q2. In a class of 50 students:
20 like Mathematics (M)
15 like Science (S)
5 like both subjects

What is the probability that a student chosen at random likes Mathematics or Science?

#### Answer :

Using the addition rule:

n(M ∪ S) = n(M) + n(S) − n(M ∩ S)

= 20 + 15 − 5

= 30

Probability = 30/50

= 3/5

= 0.6

Result: Probability = 0.6

Interpretation: There is a 60% chance that a randomly selected student likes Mathematics or Science.

Conclusion: More than half of the students like at least one of the two subjects.

### Q3. A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?

#### Answer :

Initially, the bag contains 3 red and 2 blue balls.

Given that the first ball drawn is red, the remaining balls are:

- Red balls = 2
- Total balls = 4

Probability that the next ball is red

P(Red on second draw | Red on first draw)

= 2/4

= 1/2

= 0.5

Result: Probability = 1/2

Interpretation: After removing one red ball, half of the remaining balls are red.

Conclusion: The probability that the next ball is also red is 50%.

### Q4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?

#### Answer :

Stratified Sampling should be used.

In stratified sampling, the population is divided into groups (strata) such as boys and girls, and samples are selected separately from each group.

Result: Stratified Sampling is the appropriate method.

Interpretation: This method ensures that both genders are represented according to the desired proportions.

Conclusion: Stratified Sampling is preferred because it guarantees representation from both boys and girls, whereas Simple Random Sampling may not.

### Q5. The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm. Find the sampling error.

#### Answer :

Sampling Error = Sample Mean − Population Mean

= 158 − 160

= -2 cm

Result: Sampling Error = -2 cm

Interpretation: The sample mean is 2 cm lower than the population mean.

Conclusion: The sample underestimates the population average height by 2 cm.

### Q6. The population mean salary is `₹50,000` with σ = `₹5,000`. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?

#### Answer :

Standard Error of Mean (SEM)

SEM = σ / √n

= 5000 / √100

= 5000 / 10

= 500

Result: SEM = `₹500`

Interpretation: Sample means are expected to vary by about `₹500` from the population mean.

Conclusion: The standard error of the mean is `₹500`.

### Q8. From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?

#### Answer :

Number of Aces in a deck = 4

Probability that the first card is an Ace

= 4/52

Probability that the second card is an Ace

= 3/51

P(Both Aces)

= (4/52) × (3/51)

= 12/2652

= 1/221

≈ 0.00452

Result: Probability = 1/221 ≈ 0.00452

Interpretation: The chance of drawing two Aces consecutively is very small.

Conclusion: The probability that both cards are Aces is approximately 0.452%.

### Q9. A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?

#### Answer :

Probability that a bulb is non-defective

= 1 − 0.02

= 0.98

Probability that all 5 bulbs are non-defective

= (0.98)^5

≈ 0.9039

Result: Probability ≈ 0.9039

Interpretation: There is about a 90.39% chance that all five selected bulbs are non-defective.

Conclusion: It is highly likely that all five bulbs will be non-defective.

### Q10. Differentiate between discrete and continuous random variables with examples.

#### Answer :

| Discrete Random Variable | Continuous Random Variable |
|--------------------------|----------------------------|
| Takes countable values. | Takes any value within a range. |
| Usually obtained by counting. | Usually obtained by measurement. |
| Values are distinct and separate. | Values can include decimals and fractions. |
| Example: Number of students in a class. | Example: Height of students. |
| Example: Number of cars in a parking lot. | Example: Weight of a person. |

Result: Discrete variables are countable, while continuous variables are measurable.

Conclusion: The key difference is that discrete variables take specific countable values, whereas continuous variables can take any value within an interval.
